<h3 style="color:gray;">Linear Algebra (Solving Linear Equations Using Gaussian Elimination)</h3>

```python
#This is to generate equations to test the code below.
no = int(input("Number of equations: "))
values = []
coefficients = [] #This is different from the one below. Instead of coefficients of a specific one, this is the coefficients of one equation.
results = []

for x in range(no):
    values.append(random.randrange(-10, 10))

for x in range(no):
    tempCoef = []
    for y in range(no):
        tempCoef.append(random.randrange(-25, 25))
    coefficients.append(tempCoef)
    tempRes = 0 
    for z in range(no):
        tempRes = tempRes + tempCoef[z] * values[z]
    results.append(tempRes)

print(values)
print(coefficients)
print(results)
```

```python
dimA = input(f"Enter the equation count you want to solve (Equation count must be equal to the variable count): ->")
while True:
    try:
        dimA = int(dimA)
        break
    except ValueError:
        dimA = input(f"Enter the equation count you want to solve (Equation count must be equal to the variable count): ->")

A = np.zeros((dimA, dimA + 1), dtype=np.float64)

class MatrixSingular(Exception):
    def __init__(self):
        super().__init__()
        self.message = "The determinant of the coefficient matrix is 0 (It has either none or infinite solutions.), try a different matrix."
    def __str__(self):
        return f"{self.message}"

def inputStuff():
    print(f"Please input the following values according to your linear equations in form: ax1+bx2+cx3+...+dxn=e")
    cols = []
    for x in range(dimA):
        cols.append(input(f"Enter x{x+1}'s coefficients (Leave one space between the values, if not recognized, the value will be count as 0) -> ").split(" "))
    cols.append(input(f"Enter the constants (Leave one space between the values, if not recognized, the value will be count as 0) -> ").split(" "))
    for y in range(A.shape[1]):
        for x in range(cols[y].__len__()):
            try:
                A[x, y] = float(cols[y][x])
            except ValueError:
                A[x, y] = 0
            except IndexError:
                pass

def reduceRow(row, dimension):
    if dimension <= row: return
    print(f"ELIMINATION Step {row+1}:")
    print(A)
    if row == 0: pass
    else:
        for x in range(row):
            A[row] = A[row] - A[row, x] * A[x]
    for y in range(1, A.shape[0] - 1):
        if A[row, row] == 0:
            A[row] = A[row] + A[y+1]
        else: break
    if A[row, row] == 0:
        raise MatrixSingular()
        return
    A[row] = A[row] / A[row, row]
    reduceRow(row+1, dimension)

def bsubstitute():
    for x in range(dimA - 1):
        for y in range(x + 1, dimA):
            A[x] = A[x] - A[y] * A[x, y]

def performGaussian():
    inputStuff()
    reduceRow(0, dimA)
    print(f"BACK SUBSTITUTION")
    bsubstitute()
    print(A)

performGaussian()
for x in range(dimA):
    print(f"x{x+1}={A[:,dimA][x]:.3f}",end=" ")
```

<h3 style="color:gray;">Linear Algebra (Gram-Schmidt Process for Finding Orthonormal Matrices)</h3>

You can check if the resultant matrix is orthonormal by checking the conditions: <br>
-Its column vectors should be of unit length <br>
-Each column's dot products should be zero

```python
A = np.array([[1, 16, 7, 9], #Input the matrix here. Be careful, Gram-Schmidt is only applicable to square matrices.
             [-2, 8, 1, -7],
             [6, 7, 2, -6],
             [1, 2, 8, -4]], dtype=np.float64)

A[:, 0] = A[:, 0] / np.linalg.norm(A[:, 0])
for x in range(1, A.shape[0]):
    for y in range(x):
        A[:, x] = A[:, x] - (A[:, y] @ A[:, x]) * A[:, y]
    A[:, x] = A[:, x] / np.linalg.norm(A[:, x])
A
```

<h3 style="color:gray;">Multivariate Calculus (Newton-Raphson Approximation)</h3>

```python
def func(x):
    return x**2+4*x+4

def dfuncdx(x):
    return 2*x+4

a = float(input("Enter an initial value: ->"))

steps = 0
xi = a
fxi = func(a)
dfxi = dfuncdx(a)
print(xi, fxi, dfxi)

while round(fxi, 12) != 0:
    xi = xi - (fxi/dfxi)
    fxi = func(xi)
    dfxi = dfuncdx(xi)
    print(xi, fxi, dfxi, sep=", ")
    steps += 1

print(steps)
```
